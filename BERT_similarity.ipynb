```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Huaweı\Downloads\Burak_restoran_yorumlar_ozetli.csv")
df.head()
```

```python
# bert_similarity.py
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
def calculate_bert_similarity_per_row(comments, summaries):
    model = SentenceTransformer('dbmdz/bert-base-turkish-cased')
    similarities = []
    
    for i in range(len(comments)):
        # Her raw için ayrı ayrı hesapla
        comment = comments[i]
        summary = summaries[i]
        
        # Embedding'leri hesapla
        comment_embedding = model.encode([comment])
        summary_embedding = model.encode([summary])
        
        # Cosine similarity hesapla
        similarity = cosine_similarity(comment_embedding, summary_embedding)[0][0]
        similarities.append(similarity)
    
    return similarities

# Kullanım
similarities = calculate_bert_similarity_per_row(df['Yorum'], df['Özet'])
print(f"Her raw için benzerlik skorları: {similarities}")
```

```python
# Benzerlik skorlarını DataFrame'e çevir
# Restoran isimleri ile DataFrame oluştur
similarity_df = pd.DataFrame({
    'Restoran_Ismi': df['Restoran'],  # Restoran isimleri
    'Benzerlik_Skoru': similarities
})

# İlk 5 satırı göster
print("=== İLK 5 BENZERLİK SKORU ===")
print(similarity_df.head())

# Son 5 satırı da göster
print("\n=== SON 5 BENZERLİK SKORU ===")
print(similarity_df.tail())

# En yüksek 5 benzerlik skoru
print("\n=== EN YÜKSEK 5 BENZERLİK SKORU ===")
print(similarity_df.nlargest(5, 'Benzerlik_Skoru'))

# En düşük 5 benzerlik skoru
print("\n=== EN DÜŞÜK 5 BENZERLİK SKORU ===")
print(similarity_df.nsmallest(5, 'Benzerlik_Skoru'))
```

```python
# Overall benzerlik hesaplama
def calculate_overall_similarity(similarities):
    """
    Ayrı ayrı benzerlik skorlarından overall istatistikler hesapla
    """
    overall_stats = {
        'mean_similarity': np.mean(similarities),
        'median_similarity': np.median(similarities),
        'std_similarity': np.std(similarities),
        'min_similarity': np.min(similarities),
        'max_similarity': np.max(similarities),
        'total_restaurants': len(similarities)
    }
    
    return overall_stats

# Overall benzerlik hesapla
overall_stats = calculate_overall_similarity(similarities)

print("=== OVERALL BENZERLİK İSTATİSTİKLERİ ===")
print(f"Toplam Restoran Sayısı: {overall_stats['total_restaurants']}")
print(f"Ortalama Benzerlik: {overall_stats['mean_similarity']:.4f}")
print(f"Medyan Benzerlik: {overall_stats['median_similarity']:.4f}")
print(f"Standart Sapma: {overall_stats['std_similarity']:.4f}")
print(f"Minimum Benzerlik: {overall_stats['min_similarity']:.4f}")
print(f"Maksimum Benzerlik: {overall_stats['max_similarity']:.4f}")

# Benzerlik dağılımı
print(f"\nBenzerlik Dağılımı:")
print(f"0.0-0.2: {sum(1 for x in similarities if 0.0 <= x < 0.2)} restoran")
print(f"0.2-0.4: {sum(1 for x in similarities if 0.2 <= x < 0.4)} restoran")
print(f"0.4-0.6: {sum(1 for x in similarities if 0.4 <= x < 0.6)} restoran")
print(f"0.6-0.8: {sum(1 for x in similarities if 0.6 <= x < 0.8)} restoran")
print(f"0.8-1.0: {sum(1 for x in similarities if 0.8 <= x <= 1.0)} restoran")
```

0.0-0.2: 0 restoran    → 0 restoran çok düşük benzerlik
0.2-0.4: 0 restoran    → 0 restoran düşük benzerlik
0.4-0.6: 0 restoran    → 0 restoran orta benzerlik
0.6-0.8: 53 restoran    → 53 restoran yüksek benzerlik
0.8-1.0: 2367 restoran    → 2367 restoran çok yüksek benzerlik

Cosine Similarity değerleri:
1.0 = Tamamen aynı (100% benzer)
0.0 = Hiç benzer değil (0% benzer)
0.5 = Orta düzey benzerlik (50% benzer)
